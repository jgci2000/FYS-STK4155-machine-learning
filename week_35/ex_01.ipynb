```python
import numpy as np
import numpy.linalg as linalg

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
def gen_X(x, order=1):
    X = np.zeros((len(x), order+1))
    for n in range(order+1):
        X[:, n] = np.power(x[:, 0], n)
    
    return X

def R2(y_data, y_model):
    return 1 - np.sum((y_data - y_model) ** 2) / np.sum((y_data - np.mean(y_data)) ** 2)

def MSE(y_data,y_model):
    return np.sum((y_data-y_model)**2)/np.size(y_model)
```

```python
# Setting up the data

np.random.seed(100)

N = 1000
deg = 2

x = np.random.rand(N, 1)
y = 2.0+5*x*x+0.1*np.random.randn(N, 1)

X = gen_X(x, deg)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Using OLS method

OLS_train = lambda X, y : linalg.inv(X.T @ X) @ X.T @ y
beta = OLS_train(X_train, y_train)

y_trained = X_train @ beta
y_predicted = X_test @ beta

print(f"MSE: \n - Trained: {MSE(y_train, y_trained):.2f}; "+
      f"\n - Tested: {MSE(y_test, y_predicted):.2f}")
print(f"R2 score: \n - Trained: {R2(y_train, y_trained):.2f}; "+
      f"\n - Tested: {R2(y_test, y_predicted):.2f}")
```

```python
# Using Ridge regression

n_lambdas = 20
lambdas = np.logspace(-3, 1, n_lambdas)

R2_ridge = np.zeros((2, n_lambdas))
MSE_ridge = np.zeros((2, n_lambdas))

for i, lmd in enumerate(lambdas):
    ridge_train = lambda X, y, lmd : linalg.inv(X.T @ X + lmd * np.eye(deg+1)) @ X.T @ y
    beta = ridge_train(X_train, y_train, lmd)
    
    y_trained = X_train @ beta
    y_predicted = X_test @ beta
    
    MSE_ridge[0, i] = MSE(y_train, y_trained)
    MSE_ridge[1, i] = MSE(y_test, y_predicted)
    
    R2_ridge[0, i] = R2(y_train, y_trained)
    R2_ridge[1, i] = R2(y_test, y_predicted)
```

```python
# Using Lasso regression

R2_lasso = np.zeros((2, n_lambdas))
MSE_lasso = np.zeros((2, n_lambdas))

for i, lmd in enumerate(lambdas):
    lasso_train = Lasso(alpha=lmd)
    lasso_train.fit(X_train, y_train)
    
    y_trained = lasso_train.predict(X_train)
    y_predicted = lasso_train.predict(X_test)
    
    MSE_lasso[0, i] = MSE(y_train, y_trained)
    MSE_lasso[1, i] = MSE(y_test, y_predicted)
    
    R2_lasso[0, i] = R2(y_train, y_trained)
    R2_lasso[1, i] = R2(y_test, y_predicted)

```

```python
plt.figure(1)
plt.plot(np.log10(lambdas), MSE_ridge[0, :], label="Ridge train")
plt.plot(np.log10(lambdas), MSE_ridge[1, :], label="Ridge predicted")
plt.plot(np.log10(lambdas), MSE_lasso[0, :], label="Lasso train")
plt.plot(np.log10(lambdas), MSE_lasso[1, :], label="Lasso predicted")
plt.title(r"$MSE$")
plt.xlabel(r"$\log_{10} (\lambda)$")
plt.ylabel(r"$MSE$")
plt.legend()

plt.figure(2)
plt.plot(np.log10(lambdas), R2_ridge[0, :], label="Ridge train")
plt.plot(np.log10(lambdas), R2_ridge[1, :], label="Ridge predicted")
plt.plot(np.log10(lambdas), R2_lasso[0, :], label="Lasso train")
plt.plot(np.log10(lambdas), R2_lasso[1, :], label="Lasso predicted")
plt.title(r"$R^2$ scores")
plt.xlabel(r"$\log_{10} (\lambda)$")
plt.ylabel(r"$R^2 score$")
plt.legend()

plt.show()
```
